```python
import scipy.stats as stat
import pylab 

#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist(bins=50)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.title(feature)
    plt.show()
```

```python
for feature in data.columns:
    plot_data(data, feature)
```

```python
# converting outliers to nan values
df2.loc[df2['RM']>4.173639206427822,'RM']=np.NaN
df2.loc[df2['RM']<0.5682911273305544,'RM']=np.NaN
```

```python
# filling the nan values with random sample imputation method
random_sample=df2['RM'].dropna().sample(df2['RM'].isnull().sum(),random_state=0, replace=True)

##pandas need to have same index in order to merge the dataset
random_sample.index=df2[df2['RM'].isnull()].index
df2.loc[df2['RM'].isnull(),'RM']=random_sample
```

```python
X = df2.drop('Price', axis=1)
y = df2['Price']
```

```python
# transforming independent feature using minmaxsaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit_transform(X)
```

```python
# Transforming dependent feature to log transformation
lot_y = np.log(y)
```

```python
# splitting Data into Train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaler, lot_y, test_size=0.20, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
linear = LinearRegression()
ridge = Ridge()
lasso = Lasso()
```

```python
# fitting the training Data in the model
linear.fit(X_train, y_train)
ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)
```

```python
# predicting the output
linear_pred = linear.predict(X_test)
ridge_pred = ridge.predict(X_test)
lasso_pred = ridge.predict(X_test)
```

```python
# Model evaluation
from sklearn.metrics import r2_score
linear_score = r2_score(linear_pred, y_test)
ridge_score = r2_score(ridge_pred, y_test)
lasso_score = r2_score(lasso_pred, y_test)
```
